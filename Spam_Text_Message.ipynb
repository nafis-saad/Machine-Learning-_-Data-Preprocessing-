```python
# Converting all strings to lower case
df['Message'] = df['Message'].apply(lambda x: x.lower())
df.head()
```

```python
# Removing all punctiations from messages with using regex Because:
#  same words with different punctiations may be counted 
# as different features. solo punctiations are not discriminative features.
import re
df['Message'] = df['Message'].apply(lambda x: re.sub('[^\w\s]','',x))
df.head()
```

```python
# Removing all numeric characters from Messages
df['Message'] = df['Message'].apply(lambda x:re.sub('[\d]','',x))
df.head()
```

```python
# Removing space characters more than one from Messages
df['Message'] = df['Message'].apply(lambda x:re.sub('\s\s+','',x))
df.head()
```

```python
data= df[['Category', 'Message']]
data.head()
```

```python
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize

data['Message']= data['Message'].apply(word_tokenize)
```

```python
data.head()
```

```python
#importing nlp library
import nltk
nltk.download('stopwords')
#Stop words present in the library
stopwords = nltk.corpus.stopwords.words('english')
```

```python
#defining the function to remove stopwords from tokenized text
def remove_stopwords(text):
    output= [i for i in text if i not in stopwords]
    return output
```

```python
#applying the function
data['Message']= data['Message'].apply(lambda x:remove_stopwords(x))
data.head()
```

```python
#importing the Stemming function from nltk library
from nltk.stem.porter import PorterStemmer
#defining the object for stemming
porter_stemmer = PorterStemmer()
```

```python
#defining a function for stemming
def stemming(text):
  stem_text = [porter_stemmer.stem(word) for word in text]
  return stem_text
data['Message'] = data['Message'].apply(lambda x: stemming(x))
data.head()
```

```python
from nltk.stem import WordNetLemmatizer
#defining the object for Lemmatization
wordnet_lemmatizer = WordNetLemmatizer()
```

```python
#defining the function for lemmatization
nltk.download('wordnet')
def lemmatizer(text):
 lemm_text = [wordnet_lemmatizer.lemmatize(word) for word in text]
 return lemm_text
data['Message']=data['Message'].apply(lambda x:lemmatizer(x))
```
